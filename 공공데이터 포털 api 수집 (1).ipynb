```python
import requests 
import xmltodict
import json
import pandas as pd
```

```python
response = requests.get(url,params)
response
```

```python
content = response.text
```

```python
# xml 형식의데이터
content
```

```python
# XML -> JSON 변환
# xmltodict
xml_parse = xmltodict.parse(content)
xml_dict = json.loads(json.dumps(xml_parse))
```

```python
xml_dict
```

```python
items = xml_dict['response']['result']['item']
items
```

```python
items[0]
```

```python
items[1]['goodId']
```

```python
len(items)
```

```python
goodId=[]
goodName=[]
productEntpCode=[]

for item in items:  
    if 'goodId' in item:
        goodId.append(item['goodId'])
    else:
        goodId.append(None)
    if 'goodName' in item:
        goodName.append(item['goodName'])   
    else:
        goodName.append(None)  
    if 'productEntpCode' in item:
        productEntpCode.append(item['productEntpCode'])  
    else:
        productEntpCode.append(None)
```

```python
productEntpCode
```

```python
product = pd.DataFrame({'goodId':goodId,
                        'goodName':goodName,
                        'productEntpCode':productEntpCode})
```
